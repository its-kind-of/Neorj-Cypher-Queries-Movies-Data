```python
# Query to find the title of the movies
q = 'MATCH (films:Movie)RETURN films.title LIMIT 5'

nodes = session.run(q)

for node in nodes :
    print(node)
```

```python
# Query to find the movie with the title 'The Matrix'
q = 'MATCH (matrix{title:"The Matrix"}) RETURN matrix'

nodes = session.run(q)

for node in nodes :
    print(node)
```

```python
# Query to find the movieS in which Keanu Reeves has been acted '

q = "MATCH (actor:Person)-[:ACTED_IN]-(film:Movie) WHERE actor.name='Keanu Reeves'  RETURN film.title LIMIT 5"

nodes = session.run(q)

for node in nodes :
    print(node)
```

```python
# Query to get the list of movies your favorite movies cast has been contributed 

cypher_query = '''
MATCH (movie:Movie {title:$favorite})<-[:ACTED_IN]-(actor)-[:ACTED_IN]->(rec:Movie)
 RETURN distinct rec.title as title LIMIT 20
'''
favorite_movie = input('Enter Your favorite movie to get you favorite casts movies list')
with moviesdb.session(database="neo4j") as session:
  results = session.read_transaction(
    lambda tx: tx.run(cypher_query, favorite=favorite_movie).data())
  for record in results:
    print(record['title'])
```
